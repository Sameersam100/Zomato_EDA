# Zomato Dataset Columns

18. **Aggregate Rating:** The overall rating of the restaurant.
19. **Rating Color:** A color associated with the rating.
20. **Rating Text:** Textual description of the rating.
21. **Votes:** The number of votes the restaurant has received.

# **Research Question 8: What is the geographical distribution of high-rated (rating >= 4.5) restaurants, and how does it vary across different cuisines?**

```python
import folium
from folium.plugins import MarkerCluster

# Filtering high-rated restaurants
high_rated = cleaned_df[cleaned_df['Aggregate rating'] >= 4.5]

# Creating a base map
base_map = folium.Map(location=[high_rated['Latitude'].mean(), high_rated['Longitude'].mean()], zoom_start=12)

# Creating a Marker Cluster
marker_cluster = MarkerCluster().add_to(base_map)

# Adding markers to the map
for idx, row in high_rated.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']],
                  popup=f"{row['Restaurant Name']}, Cuisine: {row['Cuisines']}, Rating: {row['Aggregate rating']}",
                  icon=folium.Icon(color='blue', icon='info-sign')).add_to(marker_cluster)

# Displaying the map
base_map
```

# **What is the relationship between the number of cuisines offered and the average rating of restaurants in each city? How does this vary with the average cost for two?**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Function to count the number of cuisines, handling non-string values
def count_cuisines(cuisines):
    if isinstance(cuisines, str):
        return len(cuisines.split(', '))
    return 0  # Return 0 if cuisines is not a string

# Creating a new column for the number of cuisines
cleaned_df['Number of Cuisines'] = cleaned_df['Cuisines'].apply(count_cuisines)

# Grouping by city and calculating the mean rating and average cost, and the average number of cuisines
city_grouped = cleaned_df.groupby('City').agg({'Aggregate rating': 'mean',
                                               'Average Cost for two': 'mean',
                                               'Number of Cuisines': 'mean'}).reset_index()

# Creating a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=city_grouped, x='Number of Cuisines', y='Aggregate rating', size='Average Cost for two', hue='City', alpha=0.7, sizes=(20, 200))
plt.title('Relationship Between Number of Cuisines, Average Rating, and Cost in Different Cities')
plt.xlabel('Average Number of Cuisines per Restaurant')
plt.ylabel('Average Aggregate Rating')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```


*This scatter plot visualizes the relationship between the number of cuisines offered, the average rating, and the average cost for two at restaurants in various cities. The color and size of the points vary according to the average cost for two, providing a multi-dimensional view of the data. The plot is generated using Plotly, a Python graphing library, as shown in the accompanying code snippet.*


**Using Plotly to make it interactive**

```python
import plotly.express as px

# Function to count the number of cuisines, handling non-string values
def count_cuisines(cuisines):
    if isinstance(cuisines, str):
        return len(cuisines.split(', '))
    return 0  # Return 0 if cuisines is not a string

# Creating a new column for the number of cuisines
cleaned_df['Number of Cuisines'] = cleaned_df['Cuisines'].apply(count_cuisines)

# Grouping by city and calculating the mean rating and average cost, and the average number of cuisines
city_grouped = cleaned_df.groupby('City').agg({'Aggregate rating': 'mean',
                                               'Average Cost for two': 'mean',
                                               'Number of Cuisines': 'mean'}).reset_index()

# Creating an interactive scatter plot
fig = px.scatter(city_grouped, x='Number of Cuisines', y='Aggregate rating',
                 size='Average Cost for two', color='City',
                 hover_name='City', size_max=60,
                 title='Relationship Between Number of Cuisines, Average Rating, and Cost in Different Cities')

fig.update_layout(xaxis_title='Average Number of Cuisines per Restaurant',
                  yaxis_title='Average Aggregate Rating',
                  legend_title='City')

fig.show()
```

The visualization helps in identifying geographical patterns in restaurant ratings.*
